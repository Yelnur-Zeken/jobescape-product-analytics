# Core Product Metrics — EDA

⚠️ **COST SAFETY**: 10 GB hard cap per query, 14-day default window. See `01_data_cleaning.ipynb` for details.

```python
from google.cloud import bigquery
import pandas as pd
import matplotlib.pyplot as plt

PROJECT = 'hopeful-list-429812-f3'
MAX_BYTES = 10 * 1024**3  # 10 GB hard cap

client = bigquery.Client(project=PROJECT)
job_config = bigquery.QueryJobConfig(maximum_bytes_billed=MAX_BYTES, use_query_cache=True)

def run(sql):
    return client.query(sql, job_config=job_config).to_dataframe()
```

## Retention cohorts (uses sql/03_retention_cohorts.sql, 14-day window)

```python
ret = run(open('../sql/03_retention_cohorts.sql').read())
ret.head()
```

```python
agg = ret.groupby('cohort_date')[['retention_d1','retention_d3','retention_d7']].mean()
ax = agg.plot(figsize=(10,4), marker='o')
ax.set_title('Avg retention by subscribe cohort (D1 / D3 / D7)')
ax.set_ylabel('rate')
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
