# Core Product Metrics — EDA

Pulls the materialized metric tables from `analytics_draft` and produces baseline charts:
- DAU / WAU / MAU
- Retention D1 / D3 / D7 over time
- Onboarding completion rate by platform

```python
from google.cloud import bigquery
import pandas as pd
import matplotlib.pyplot as plt

client = bigquery.Client(project='hopeful-list-429812-f3')
```

```python
ret = client.query(open('../sql/03_retention_cohorts.sql').read()).to_dataframe()
ret.set_index('cohort_date')[['retention_d1','retention_d3','retention_d7']].plot(figsize=(10,4))
plt.title('Retention by install cohort')
plt.ylabel('rate')
plt.show()
```
